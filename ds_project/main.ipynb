```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LassoCV
import pickle

dfam = pd.read_csv('CO2 Emissions_Canada.csv')

dfam.replace(to_replace= 'Z', value= 'Premium gasoline', inplace = True)
dfam.replace(to_replace= 'X', value= 'Gasoline' , inplace= True)
dfam.replace(to_replace= 'D', value= 'Diesel' , inplace= True)
dfam.replace(to_replace= 'E', value= 'Ethanol' , inplace= True)
dfam.replace(to_replace= 'N', value= 'Natural gas' , inplace= True)

dfam.drop(['Transmission', 'Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)'], axis = 1, inplace = True)
# oh, forgot to update the df, yep
dfam.head()
```

```python
col_category_ameya = ['Make', 'Model', 'Vehicle Class' , 'Fuel Type']
for col in col_category_ameya:
    dfam = pd.concat([dfam.drop(col, axis=1), pd.get_dummies(dfam[col], prefix=col, prefix_sep='_')], axis=1)

#LOLLLLLLL we got an extra square bracket ok im glad that it works
# yea :/
input_columns = dfam.loc[:, dfam.columns != 'CO2 Emissions(g/km)']
target = dfam["CO2 Emissions(g/km)"]
x_trainam, x_testam, y_trainam, y_testam = train_test_split(input_columns, target, test_size=0.2)


lassoam = LassoCV()
lassoam.fit(x_trainam, y_trainam)
y_hatam = lassoam.predict(x_testam)

from sklearn.metrics import mean_squared_error
from math import sqrt

#calculate MSE
total_squared_error = (np.sum((y_testam - y_hatam)**2)) #get the sum of all the errors (error = what we want (y_test) - what we predicted (y_hat))
mean_squared_error = total_squared_error/len(y_testam) #divide this by how many rows/observations we have 
print(mean_squared_error)

#calculate RMSE
rmse = sqrt(mean_squared_error)
print(rmse)

# normalize it
normrmse = rmse/(max(dfam['CO2 Emissions(g/km)'])-min(dfam['CO2 Emissions(g/km)']))
print(normrmse)

# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(lassoam, open(filename, 'wb'))
```
