### Exploratory Data Analysis

```python
# create plots for your data to understand the trend of data :D
```

```python
#Michael
#X- Engine Size, Y- CO2 Emissions, Color- Type of Fuel
#Trend: as the size of the engine increases so does the CO2 emissions

fig = px.scatter(co2_dataframe, x= 'Engine Size(L)', y= 'CO2 Emissions(g/km)', color='Fuel Type', title= 'CO2 Emissions by Engine Size')

#Changes the data's fuel label (X,Z,D...) to its actual fuel name
newnames = {'X': 'Regular gasoline', 'Z': 'Premium gasoline', 'D': 'Diesel', 'E': 'Ethanol (E85)', 'N': 'Natural gas'} #{'data's fuel name': 'Actual fuel name'}
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
fig.show()
```

```python
#Michael
#X- Vehicle Class, Y- CO2 Emissions, Color- Type of Fuel
#Trend: There isn't much of a trend between vehicle class and CO2 Emissions, emissions might be more dependent on the engine

fig = px.scatter(co2_dataframe, x= 'Vehicle Class', y= 'CO2 Emissions(g/km)', color='Fuel Type', size= 'Engine Size(L)', title = 'CO2 Emissions by Vehicle class')

#Changes the data's fuel label (X,Z,D...) to its actual fuel name
newnames = {'X': 'Regular gasoline', 'Z': 'Premium gasoline', 'D': 'Diesel', 'E': 'Ethanol (E85)', 'N': 'Natural gas'} #{'data's fuel name': 'Actual fuel name'}
fig.for_each_trace(lambda t: t.update(name = newnames[t.name], legendgroup = newnames[t.name], hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])))

fig.show()
```

```python
#Michael
#X- Vehicle Make, Y- CO2 Emissions, Color- Type of Fuel
#Trend: There isn't much of a trend between vehicle class and CO2 Emissions, emissions might be more dependent on the engine

fig = px.scatter(co2_dataframe, x= 'Make', y= 'CO2 Emissions(g/km)', color='Vehicle Class', title = 'CO2 Emissions by Vehicle class')
fig.show()
```
