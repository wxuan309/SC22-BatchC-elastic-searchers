### Exploratory Data Analysis



```python
# create plots for your data to understand the trend of data :D
```

```python
fig = px.scatter(df, x="Engine Size(L)", y="CO2 Emissions(g/km)", color="Fuel Type", title="CO2 Emissions by Engine Size")
fig.show()
# shows the CO2 emissions based on the engine size. general trend towards the bigger the engine, the higher CO2 emissions
```

```python
fig1 = px.scatter(df, x="Cylinders", y="CO2 Emissions(g/km)", color="Fuel Type", title="CO2 Emissions by Cylinders")
fig1.show()
# shows the CO2 emissions based on the amount of cylinders. general trend towards the more cylinders, the higher CO2 emissions
```

```python
fig3 = px.scatter(df, x="Make", y="CO2 Emissions(g/km)", color="Fuel Type", title="CO2 Emissions by Make")
fig3.show()
# shows the CO2 emissions based on the Make. no general trend, but interesting to see which brands have generally lower emissions (Fiat, Smart, etc.) and which ones have generally higher emissions (Bugatti, Lamborghini, etc.)
```

```python
fig2 = px.scatter(df, x="Fuel Consumption Comb (mpg)", y="CO2 Emissions(g/km)", color="Fuel Type", title="CO2 Emissions by Fuel Consumption")
fig2.show()
# shows the CO2 emissions based on the Fuel Consumption. general exponential decay where the larger the mpg, the fewer emissions. each fuel type along a pretty clear line
```
